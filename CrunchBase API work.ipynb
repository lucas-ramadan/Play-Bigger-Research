### Use Regex to clean the description of Hotlinks

```python
link_pattern = '\[[\w ]*\]\([\w+]*[://]*[w]*[/.][\w]*[/.][/\w-]*\)'
```

```python
fb_cleaned = re.sub(link_pattern, '', fb_desc)
```

```python
fb_cleaned
```

```python
from nltk.tokenize import word_tokenize
import string
import ftfy
```

```python
# we notice token facebook\u2019s, which should really be facebook's
```

```python
u'facebook\u2019s'
```

```python
ftfy.ftfy(u'facebook\u2019s')
```

```python
fb_tokens = word_tokenize(ftfy.ftfy(fb_cleaned))
```

```python
fb_tokens
```

```python
clean_fb_tokens = [word.lower() for word in fb_tokens if word not in string.punctuation]
```

```python
clean_fb_tokens
```

### Iterate through ODM, get list of all the companies that were founded since 2000

```python
odm = pd.read_csv('data/odm2.csv')
```

```python
odm.head()
```

```python
# how many entity names are there?
entity_names = odm['name']
```

```python
len(entity_names)
```

```python
len(set(entity_names))
```

```python
# how many repeats do there appear to be? ~46K 
seen = set()
copies = set()

for comp in entity_names:
    if comp not in seen:
        seen.add(comp)
    else:
        copies.add(comp)

print len(seen), len(copies)
```

```python
# only two "types" of entities, Organizations and Persons
print set(odm['type'])
```

```python
from collections import Counter
```

```python
# keep just the organizations 
orgs = odm[odm['type'] == 'Organization']
```

```python
len(orgs)
```

```python
# we want only the companies
Counter(orgs['primary_role']).most_common(10)
```

```python
companies = orgs[orgs['primary_role'] == 'company']
```

```python
us_companies = companies[companies['location_country_code']=='USA']
```

```python
len(us_companies)
```
